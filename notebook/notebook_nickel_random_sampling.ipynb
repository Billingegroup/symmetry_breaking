# Random Sampling

```python
import sys
sys.path.append('../')
from symmetry_breaking_measure import Nickel
import matplotlib.pyplot as plt
from bg_mpl_stylesheet.bg_mpl_stylesheet import bg_mpl_style
from colour import Color
from matplotlib.colors import LinearSegmentedColormap

def make_Ramp(ramp_colors): 
    color_ramp = LinearSegmentedColormap.from_list( 'my_list', [ Color( c1 ).rgb for c1 in ramp_colors ] )
    return color_ramp

def rgb_to_hex(rgb):
    return '#{:02x}{:02x}{:02x}'.format(int(rgb[0]*255), int(rgb[1]*255), int(rgb[2]*255))

MIDDLE, RED, BLUE = '#a8b6c1', '#B82601', '#0B3C5D'
NUM_SAMPLES = 20000
cmap_blue = make_Ramp([BLUE, MIDDLE, BLUE])
cmap_red = make_Ramp([MIDDLE, RED])
cmap_red = make_Ramp([rgb_to_hex(cmap_red(cmap_red.N//2)[:3]), RED])
```

```python
# Spheroid cut off from Nickel (radius=1)
spheroid_nickel = Nickel(cutoff='ellipsoid', cutout_x=1, cutout_y=1, cutout_z=1)
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
plt.style.use(bg_mpl_style)

ax = spheroid_nickel.plot_ellipsoid(ax=ax, num_samples=NUM_SAMPLES, cmap=cmap_blue)
ax.set_xlim(-4.2,4.2)
ax.set_ylim(-4.1,4.1)
ax.set_zlim(-3.8,3.8)
ax.set_axis_off()

plt.show()
#fig_directory = '../figs/nickel/spheroid_nickel_samples.png'
fig_directory = '../figs/nickel/spheroid_nickel_samples_nobackground.png'
fig.savefig(fig_directory)
```

```python
# Ellipsoid cut off from Nickel
ellipsoid_nickel = Nickel(cutoff='ellipsoid', cutout_x=2, cutout_y=1, cutout_z=1)
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
plt.style.use(bg_mpl_style)

ax = ellipsoid_nickel.plot_ellipsoid(ax=ax, num_samples=NUM_SAMPLES, cmap=cmap_blue)
ax.set_xlim(-7,7)
ax.set_ylim(-7,7)
ax.set_zlim(-7,7)
ax.set_axis_off()

plt.show()
# fig_directory = '../figs/nickel/ellipsoid_nickel_samples.png'
fig_directory = '../figs/nickel/ellipsoid_nickel_samples_nobackground.png'
fig.savefig(fig_directory)
```

```python
# Cubic solid cut off from Nickel (length=1)
nickel = Nickel(cutoff='rectangular_solid', cutout_x=1, cutout_y=1, cutout_z=1)
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
plt.style.use(bg_mpl_style)

cmap_blue = make_Ramp([MIDDLE, BLUE])
ax = nickel.plot_rectangular_solid(ax=ax, num_samples=NUM_SAMPLES, cmap=cmap_red, facecolor=RED)
ax.set_xlim(-0.2,4.5)
ax.set_ylim(-0.2,4)
ax.set_zlim(-0.2,3.7)
ax.set_axis_off()

plt.show()
#fig_directory = '../figs/nickel/cubic_solid_nickel_samples.png'
fig_directory = '../figs/nickel/cubic_solid_nickel_samples_nobackground.png'
fig.savefig(fig_directory)
```

```python
# Rectangular solid cut off from Nickel
rectangular_solid_nickel = Nickel(cutoff='rectangular_solid', cutout_x=2, cutout_y=1, cutout_z=1)
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
plt.style.use(bg_mpl_style)

ax = rectangular_solid_nickel.plot_rectangular_solid(ax=ax, num_samples=NUM_SAMPLES, cmap=cmap_red, facecolor=RED)
ax.set_xlim(-.5,7.5)
ax.set_ylim(-.5,5)
ax.set_zlim(-.5,4.7)
ax.set_axis_off()

plt.show()
# fig_directory = '../figs/nickel/rectangular_solid_nickel_samples.png'
fig_directory = '../figs/nickel/rectangular_solid_nickel_samples_nobackground.png'
fig.savefig(fig_directory)
```
